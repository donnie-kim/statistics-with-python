2. Write a function that inputs a list of integers and returns the minimum value


````python
# Write your function here
def return_min(lst):
    minimum = lst[0]
    for i in lst:
        if i < minimum:
            minimum = i
    return minimum
````

#### Challenge problem:

Write a function that take in four arguments: lst1, lst2, str1, str2, and returns a pandas DataFrame that has the first column labeled str1 and the second column labaled str2, that have values lst1 and lst2 scaled to be between 0 and 1.

For example

```
lst1 = [1, 2, 3]
lst2 = [2, 4, 5]
str1 = 'one'
str2 = 'two'

my_function(lst1, lst2, str1, str2)
```

should return a DataFrame that looks like:

| <i></i> | one | two  |
| ------- | --- | ---- |
| 0       | 0   | 0    |
| 1       | .5  | .666 |
| 2       | 1   | 1    |


````python
import pandas as pd
import numpy as np


def return_dataframe(lst1, lst2, str1, str2):
    interp_lst1 = np.interp(lst1, (min(lst1), max(lst1)), (0, 1))
    interp_lst2 = np.interp(lst2, (min(lst2), max(lst2)), (0, 1))

    df = pd.DataFrame({str1: interp_lst1, str2: interp_lst2})
    return df
````
